```python
"""
HealthRisk AI
Feature Interaction Analysis using SHAP

Supports:
- XGBoost
- LightGBM

Stacking ensembles are intentionally not supported because SHAP
TreeExplainer cannot explain sklearn StackingRegressor directly.
"""

from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import shap


class FeatureInteractionAnalyzer:

    def __init__(self, tree_model):

        self.model = tree_model
        self.explainer = shap.TreeExplainer(tree_model)

    def interaction_values(self, X):

        return self.explainer.shap_interaction_values(X)

    def summary_plot(
        self,
        X,
        output_dir="reports",
        filename="shap_interaction_summary.png"
    ):

        output_dir = Path(output_dir)
        output_dir.mkdir(exist_ok=True)

        interaction_values = self.interaction_values(X)

        plt.figure(figsize=(10,6))

        shap.summary_plot(
            interaction_values,
            X,
            show=False
        )

        path = output_dir / filename

        plt.tight_layout()

        plt.savefig(path, dpi=300)

        plt.close()

        return path

    def dependence_plot(
        self,
        feature,
        X,
        output_dir="reports"
    ):

        output_dir = Path(output_dir)
        output_dir.mkdir(exist_ok=True)

        interaction_values = self.interaction_values(X)

        shap.dependence_plot(
            feature,
            interaction_values,
            X,
            show=False
        )

        path = output_dir / f"interaction_{feature}.png"

        plt.tight_layout()

        plt.savefig(path, dpi=300)

        plt.close()

        return path
```
