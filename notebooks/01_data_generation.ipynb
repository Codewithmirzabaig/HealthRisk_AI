```python
import numpy as np
import pandas as pd
from pathlib import Path

np.random.seed(42)

n = 10000

age = np.random.randint(18, 90, n)
bmi = np.round(np.random.normal(28, 6, n), 1)
bmi = np.clip(bmi, 16, 50)

smoker = np.random.binomial(1, 0.18, n)
diabetes = np.random.binomial(1, np.clip((age - 30) / 100, 0.05, 0.45), n)
hypertension = np.random.binomial(1, np.clip((age - 25) / 90, 0.08, 0.60), n)
heart_disease = np.random.binomial(1, np.clip((age - 45) / 120, 0.03, 0.35), n)

hospital_visits = np.random.poisson(
    1 + diabetes * 1.2 + hypertension * 0.8 + heart_disease * 1.8 + smoker * 0.7
)

medication_count = np.random.poisson(
    1 + diabetes * 2 + hypertension * 1.5 + heart_disease * 2
)

premium_paid = np.round(
    3000 + age * 45 + smoker * 1200 + diabetes * 1800 + hypertension * 900 + heart_disease * 2500,
    2
)

annual_claim_amount = np.round(
    1000 + age * 80 + bmi * 70 + smoker * 1500 + diabetes * 3000 +
    hypertension * 1800 + heart_disease * 6000 + hospital_visits * 2200 +
    np.random.normal(0, 2500, n),
    2
)
annual_claim_amount = np.clip(annual_claim_amount, 500, None)

pandemic_risk = np.random.uniform(0, 1, n)

future_health_cost = np.round(
    annual_claim_amount * np.random.uniform(1.05, 1.45, n)
    + pandemic_risk * 3000,
    2
)

risk_score = (
    age * 0.02 + bmi * 0.03 + smoker * 1.5 + diabetes * 2 +
    hypertension * 1.2 + heart_disease * 3 + hospital_visits * 0.6 +
    medication_count * 0.25 + pandemic_risk * 2
)

risk_level = pd.cut(
    risk_score,
    bins=[-1, 4.5, 8.5, 100],
    labels=["Low", "Medium", "High"]
)

df = pd.DataFrame({
    "patient_id": range(1, n + 1),
    "age": age,
    "gender": np.random.choice(["Male", "Female"], n),
    "bmi": bmi,
    "smoker": smoker,
    "diabetes": diabetes,
    "hypertension": hypertension,
    "heart_disease": heart_disease,
    "hospital_visits": hospital_visits,
    "medication_count": medication_count,
    "annual_claim_amount": annual_claim_amount,
    "premium_paid": premium_paid,
    "hospital_rating": np.random.randint(1, 6, n),
    "hospital_debt_ratio": np.round(np.random.uniform(0.15, 0.85, n), 2),
    "pharma_investment_score": np.round(np.random.uniform(0, 100, n), 2),
    "esg_score": np.round(np.random.uniform(0, 100, n), 2),
    "pandemic_risk_index": np.round(pandemic_risk, 3),
    "future_health_cost": future_health_cost,
    "risk_level": risk_level
})

output_path = r"C:\Users\Mirza Shareef Baig\OneDrive\Desktop\Healthrisk-ai\data\raw\patient_financial_risk.csv"
df.to_csv(output_path, index=False)

print(f"Dataset created successfully: {output_path}")
print(df.head())
```
